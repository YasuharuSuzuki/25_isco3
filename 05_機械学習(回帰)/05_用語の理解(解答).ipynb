### 🎲 四択問題14 過学習 (3点)

- 過学習（Overfitting）の説明として最も適切なものを次から選べ

    📕 1. モデルが訓練データに対して過度に単純である状態  
    📗 2. モデルが訓練データを完全に学習しすぎて、テストデータに対して性能が低下する状態  
    📘 3. モデルが訓練データに対して十分に学習できない状態  
    📙 4. モデルがテストデータに対して過度に複雑である状態  

```python
# 正解を出力してください
print('2')
```

---

### ✅ 解説

**過学習（Overfitting）** とは、モデルが訓練データの特徴やノイズまで **過剰に学習** してしまい、未知のデータ（テストデータ）に対して **うまく予測できなくなる状態** を指します。

#### **各選択肢の解説**
📕 1. モデルが訓練データに対して過度に単純である状態  
- 学習能力が不足している状態であり、 **過少学習（Underfitting）** の説明です。  

📗 2. モデルが訓練データを完全に学習しすぎて、テストデータに対して性能が低下する状態  
- 過学習の典型的な説明であり、正解です。  

📘 3. モデルが訓練データに対して十分に学習できない状態  
- これも **過少学習（Underfitting）** を表しています。

📙 4. モデルがテストデータに対して過度に複雑である状態  
- モデルの複雑さは訓練時に決まるものであり、表現として不適切です。  
過学習は「訓練データへの適合しすぎ」が本質です。

## 問17 性能を評価するために使用するデータ (3点)

- 学習済みモデルの性能を評価する際に使用するのはどのデータか？

    📕 1. テストデータ  
    📗 2. 訓練データ  
    📘 3. 説明変数  
    📙 4. 水増しデータ  

```python
# 正解を出力してください
print('1')
```

---

### ✅ 解説

学習済みモデルの性能評価では、モデルが **まだ見たことのないデータに対して** どれだけ正しく予測できるかを確認する必要があります。  
そのため、学習には使用していないテストデータを用いて評価を行います。  

#### **各選択肢の解説**

📕 1. テストデータ  
- 学習に使われていないデータであり、 **汎化性能（未知データへの対応力）** を評価するために使用します。
よって **正解** です。

📗 2. 訓練データ  
- モデルを学習させるためのデータであり、これで評価すると **過学習に気づけない** ため、性能評価には不適切です。

📘 3. 説明変数  
- 入力データの項目を指す用語であり、データの種類（訓練・テスト）を表すものではありません。

📙 4. 水増しデータ  
- データ拡張（Data Augmentation）によって生成されたデータであり、学習用に使われることはありますが、**性能評価用のデータではありません** 。
