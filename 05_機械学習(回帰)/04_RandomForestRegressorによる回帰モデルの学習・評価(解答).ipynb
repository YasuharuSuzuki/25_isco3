### ğŸ©º ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 11 ã‚¨ãƒ©ãƒ¼ãƒ¬ãƒ¼ãƒˆã¨æ±ºå®šä¿‚æ•°ã«ã‚ˆã‚‹ãƒ¢ãƒ‡ãƒ«ã®è©•ä¾¡

- ã‚¨ãƒ©ãƒ¼ãƒ¬ãƒ¼ãƒˆã¨æ±ºå®šä¿‚æ•°ã‚’ç®—å‡ºã—ã¦ã€ãƒ¢ãƒ‡ãƒ«ã‚’è©•ä¾¡ã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚


```python
# ã‚¨ãƒ©ãƒ¼ãƒ¬ãƒ¼ãƒˆã¨æ±ºå®šä¿‚æ•°ã®ç®—å‡ºã‚’è¡Œã†ãƒ¦ãƒ¼ã‚¶ãƒ¼å®šç¾©é–¢æ•°
def evaluate_metrics(y_train, y_test, y_train_pred, y_test_pred):
    # å­¦ç¿’æ™‚ã‚¹ã‚³ã‚¢
    train_MAE = mean_absolute_error(y_train, y_train_pred)
    train_MSE = mean_squared_error(y_train, y_train_pred)
    train_RMSE = np.sqrt(mean_squared_error(y_train, y_train_pred))
    train_R2 = r2_score(y_train, y_train_pred)
    
    # ãƒ†ã‚¹ãƒˆæ™‚ã‚¹ã‚³ã‚¢
    test_MAE = mean_absolute_error(y_test, y_test_pred)
    test_MSE = mean_squared_error(y_test, y_test_pred)
    test_RMSE = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_R2 = r2_score(y_test, y_test_pred)
    
    print("å­¦ç¿’æ™‚ã‚¹ã‚³ã‚¢(train score)")
    print("ã€€å¹³å‡çµ¶å¯¾èª¤å·®: ", train_MAE)
    print("ã€€å¹³å‡äºŒä¹—èª¤å·®: ", train_MSE)
    print("ã€€äºŒä¹—å¹³å‡å¹³æ–¹æ ¹èª¤å·®: ", train_RMSE)
    print("ã€€æ±ºå®šä¿‚æ•°R^2: ", train_R2)
    
    print("ãƒ†ã‚¹ãƒˆã‚¹ã‚³ã‚¢(test score)")
    print("ã€€å¹³å‡çµ¶å¯¾èª¤å·®: ", test_MAE)
    print("ã€€å¹³å‡äºŒä¹—èª¤å·®: ", test_MSE)
    print("ã€€äºŒä¹—å¹³å‡å¹³æ–¹æ ¹èª¤å·®: ", test_RMSE)
    print("ã€€æ±ºå®šä¿‚æ•°R^2: ", test_R2)
```

```python
y_train_tips_pred = rf_model_tips.predict(X_train_tips)  # å­¦ç¿’ãƒ‡ãƒ¼ã‚¿ã‚’ä½¿ã£ã¦æ¨è«–ã™ã‚‹
y_test_tips_pred = rf_model_tips.predict(X_test_tips)  # ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã‚’ä½¿ã£ã¦æ¨è«–ã™ã‚‹

# ã‚¨ãƒ©ãƒ¼ãƒ¬ãƒ¼ãƒˆã¨æ±ºå®šä¿‚æ•°ã®ç®—å‡ºã‚’è¡Œã†ãƒ¦ãƒ¼ã‚¶ãƒ¼å®šç¾©é–¢æ•°ã®å‘¼ã³å‡ºã—
evaluate_metrics(y_train_tips, y_test_tips, y_train_tips_pred, y_test_tips_pred)
```

```python
# æ®‹å·®ã®å¯è¦–åŒ–ã‚’è¡Œã†ãƒ¦ãƒ¼ã‚¶ãƒ¼å®šç¾©é–¢æ•°
def residual_plot(y_column_name, y_train, y_test, y_train_pred, y_test_pred):
    df_train = pd.DataFrame(y_train)
    df_train['y_train_pred'] = y_train_pred
    
    df_test = pd.DataFrame(y_test)
    df_test['y_test_pred'] = y_test_pred
    
    ax = plt.gca()
    
    # å‚¾ã1ã®ç›´ç·šã‚’è¿½åŠ 
    a_line=[min(df_train[y_column_name].min(), df_test[y_column_name].min()), max(df_train[y_column_name].max(), df_test[y_column_name].max())]
    sns.regplot(x=df_train[y_column_name], y=df_train['y_train_pred'],label='train')
    sns.regplot(x=df_test[y_column_name], y=df_test['y_test_pred'],label='test')
    sns.regplot(x=a_line, y=a_line,ci=None, scatter=False, line_kws={'color': 'orange', 'linestyle': '--', 'lw': 1},label='residual 0')
    
    ax.set_xlabel('true')
    ax.set_ylabel('pred')
    
    plt.legend(fontsize=10)
    plt.show()
```
