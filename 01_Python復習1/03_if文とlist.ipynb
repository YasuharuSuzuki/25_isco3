## â”ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 4ã€€ifæ–‡
ifæ–‡ã‚’ä½¿ã£ãŸæ¡ä»¶åˆ†å²ã‚’ç¿’å¾—ã—ã¾ã—ã‚‡ã†

```python
# ifæ–‡
# input()ã§ç‚¹æ•°ã‚’å—ã‘å–ã‚Šã€60ç‚¹æœªæº€ã®æ™‚ã¯â€ä¸å¯â€ã¨è¡¨ç¤ºã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ 
exam_score1 = int(input()) # å…¥åŠ›ã—ãŸå€¤ã‚’intã«å¤‰æ›ã—ã¦å–å¾—
if exam_score1 < 60: # 60ç‚¹æœªæº€ã‹ã©ã†ã‹ã‚’æ¯”è¼ƒã€‚æ¡ä»¶ã«é©åˆã™ã‚‹å ´åˆã¯ã€ã€ã€
    print("ä¸å¯") # â€ä¸å¯â€ã¨è¡¨ç¤ºã™ã‚‹
```

```python
# if - else æ–‡
# input()ã§ç‚¹æ•°ã‚’å—ã‘å–ã‚Šã€60ç‚¹æœªæº€ã®æ™‚ã¯â€ä¸å¯â€ã€60ç‚¹ä»¥ä¸Šã®å ´åˆã¯"å¯"ã¨è¡¨ç¤ºã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ 
exam_score2 = int(input()) # å…¥åŠ›ã—ãŸå€¤ã‚’intã«å¤‰æ›ã—ã¦å–å¾—
if exam_score2 < 60: # 60ç‚¹æœªæº€ã‹ã©ã†ã‹ã‚’æ¯”è¼ƒã€‚æ¡ä»¶ã«é©åˆã™ã‚‹å ´åˆã¯ã€
    print("ä¸å¯") # â€ä¸å¯â€ã¨è¡¨ç¤ºã™ã‚‹
else: # æ¡ä»¶ã«é©åˆã—ãªã„å ´åˆã¯
    print("å¯") # â€å¯â€ã¨è¡¨ç¤ºã™ã‚‹
```

```python
# if - elif - else æ–‡
# input()ã§ç‚¹æ•°ã‚’å—ã‘å–ã‚Šã€90ç‚¹ä»¥ä¸Šã®æ™‚ã¯â€ç§€â€ã€80ç‚¹ä»¥ä¸Šã®å ´åˆã¯"å„ª"ã€70ç‚¹ä»¥ä¸Šã®å ´åˆã¯"è‰¯"ã€60ç‚¹ä»¥ä¸Šã®å ´åˆã¯"å¯"ã€
# 60ç‚¹æœªæº€ã®å ´åˆã¯"ä¸å¯"ã¨è¡¨ç¤ºã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ 
exam_score3 = int(input())
if exam_score3 >= 90:
    print("ç§€")
elif exam_score3 >= 80:
    print("å„ª")
elif exam_score3 >= 70:
    print("è‰¯")
elif exam_score3 >= 60:
    print("å¯")
else:
    print("ä¸å¯")
```
