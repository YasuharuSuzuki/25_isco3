```python
# ãƒ‡ãƒ¼ã‚¿ãƒ•ã‚¡ã‚¤ãƒ«ãƒ‘ã‚¹å®šç¾©
WINE_CSV = "./data/wine.csv"
IRIS_CSV = "./data/iris.csv"

# ãƒ¢ãƒ‡ãƒ«ãƒ•ã‚¡ã‚¤ãƒ«ãƒ‘ã‚¹å®šç¾©
WINE_SVC_MODEL = "./data/wine_svc.joblib"
IRIS_SVC_MODEL = "./data/iris_svc.joblib"

# RANDOM SEED å®šç¾©
RANDOM_SEED = 0
```

### ğŸ¾ ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 1 ãƒ¯ã‚¤ãƒ³åˆ†é¡ãƒ¢ãƒ‡ãƒ«ã®å­¦ç¿’

```python
# ãƒ¯ã‚¤ãƒ³ã®ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã‚’èª­ã¿è¾¼ã¿
df_wine = pd.read_csv(WINE_CSV, index_col=None) # indexã«ä½¿ç”¨ã™ã‚‹ã‚«ãƒ©ãƒ ãŒç„¡ã„ãŸã‚ã€index_col=Noneã‚’æŒ‡å®šã—ã¾ã™
```

```python
# SVCã§åˆ†é¡ãƒ¢ãƒ‡ãƒ«ã‚’ä½œã‚‹
# èª¬æ˜å¤‰æ•°ã¨ç›®çš„å¤‰æ•°ã«åˆ†ã‘ã‚‹
X = df_wine.iloc[:,:-1]
y = df_wine.iloc[:,-1]
```

```python
# è¨“ç·´ãƒ‡ãƒ¼ã‚¿ã¨ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã«åˆ†å‰²ã™ã‚‹
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=RANDOM_SEED) # ç‰¹ã«ç†ç”±ãŒãªã„é™ã‚Šã¯ã€å­¦ç¿’ãƒ‡ãƒ¼ã‚¿ï¼šãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã¯ 7ï¼š3 ã§ã‚ã‘ã¾ã—ã‚‡ã†ã€‚
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=RANDOM_SEED, stratify=y)  # stratifyã‚’æŒ‡å®šã™ã‚‹ã¨å±¤åˆ¥ã«ãã‚Œã„ã«åˆ†ã‘ã‚‹ã“ã¨ãŒã§ãã¾ã™
```

```python
# å­¦ç¿’ã™ã‚‹
svc_model = svm.SVC()    # ãƒ¢ãƒ‡ãƒ«ã‚’ä½œã‚‹
svc_model.fit(X_train, y_train)    # è¨“ç·´ã™ã‚‹
```

```python
# å­¦ç¿’ã—ãŸãƒ¢ãƒ‡ãƒ«ã‚’ä¿å­˜ã™ã‚‹
joblib.dump(svc_model, WINE_SVC_MODEL)
```
