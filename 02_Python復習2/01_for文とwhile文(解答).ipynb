## ğŸ’« ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 1ã€€foræ–‡
foræ–‡ã®ä½¿ã„æ–¹ã‚’ç¿’å¾—ã—ã¾ã—ã‚‡ã†
- ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™
- ãƒªã‚¹ãƒˆã®è¦ç´ ã‚’é †ç•ªã«å–ã‚Šå‡ºã™
- ãƒªã‚¹ãƒˆã®è¦ç´ æ•°ã ã‘ç¹°ã‚Šè¿”ã™
- ãƒªã‚¹ãƒˆã®è¦ç´ æ•°ã ã‘ç¹°ã‚Šè¿”ã™2
- ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã§ã®foræ–‡
- ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã‚’ä½¿ã‚ãšã«åŒã˜ãƒªã‚¹ãƒˆã‚’ä½œã‚‹

```python
# ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™
N = 5 # ç¹°ã‚Šè¿”ã—ãŸã„æ•°
for index in range(N):
    print(index)
```

```python
# ãƒªã‚¹ãƒˆã®è¦ç´ ã‚’é †ç•ªã«å–ã‚Šå‡ºã™
fruits = ["apple", "banana", "orange"]
for item in fruits:
    print(item) # apple, banana, orange
```

```python
# ãƒªã‚¹ãƒˆã®è¦ç´ æ•°ã ã‘ç¹°ã‚Šè¿”ã™
list1 = [4, 5, 2, 6, 3, 2] # ä»»æ„ã®é…åˆ—
for index in range(len(list1)): # é…åˆ—è¦ç´ æ•°
    print(index) # 0, 1, 2, 3, 4, 5
    print(list1[index]) # 4, 5, 2, 6, 3, 2
```

```python
# ãƒªã‚¹ãƒˆã®è¦ç´ æ•°ã ã‘ç¹°ã‚Šè¿”ã™2
list1 = [4, 5, 2, 6, 3, 2] # ä»»æ„ã®é…åˆ—
for index, item in enumerate(list1):
    print(index) # 0, 1, 2, 3, 4, 5
    print(item) # 4, 5, 2, 6, 3, 2
```

```python
# ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã§ã®foræ–‡
list1 = [1, 2, 3, 4, 5] # ä»»æ„ã®é…åˆ—
list5 = [num1*5 for num1 in list1]
print(list5)
```

```python
# ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã‚’ä½¿ã‚ãšã«åŒã˜ãƒªã‚¹ãƒˆã‚’ä½œã‚‹
list1 = [1, 2, 3, 4, 5] # ä»»æ„ã®é…åˆ—
list5 = []
for num1 in list1:
    list5.append(num1*5)
print(list5) # [5, 10, 15, 20, 25]
```

2ï¸âƒ£ ãƒªã‚¹ãƒˆã®è¦ç´ ã‚’é †ç•ªã«å–ã‚Šå‡ºã—ã¦å‡ºåŠ›ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†  
3ï¸âƒ£ enumerate()ã‚’ä½¿ã£ã¦ã€ãƒªã‚¹ãƒˆã®ä½ç½®ã¨è¦ç´ ã‚’ä¸¡æ–¹å‡ºåŠ›ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†  
4ï¸âƒ£ ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã§foræ–‡ã‚’ä½¿ã£ã¦ã¿ã¾ã—ã‚‡ã†  

```python
# 1. ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†
# å‡ºåŠ›ã™ã‚‹å€¤ã¯æ–‡å­—åˆ—ã§ã‚‚æ•´æ•°å€¤ã§ã‚‚æµ®å‹•å°æ•°å€¤ã§ã‚‚ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’å…¥ã‚Œã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

repeat = 7
for index in range(0,repeat,2):
    print(index)
```

```python
# 2. ãƒªã‚¹ãƒˆã®è¦ç´ ã‚’é †ç•ªã«å–ã‚Šå‡ºã—ã¦å‡ºåŠ›ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†
# å‡ºåŠ›ã™ã‚‹å€¤ã¯æ–‡å­—åˆ—ã§ã‚‚æ•´æ•°å€¤ã§ã‚‚æµ®å‹•å°æ•°å€¤ã§ã‚‚ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’å…¥ã‚Œã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

japan = ["åŒ—æµ·é“","æ±åŒ—","é–¢æ±","ä¸­éƒ¨","è¿‘ç•¿","ä¸­å›½","å››å›½","ä¹å·"]
for japan_chiho in japan:
    print(japan_chiho)
```

```python
# 3. enumerate()ã‚’ä½¿ã£ã¦ã€ãƒªã‚¹ãƒˆã®ä½ç½®ã¨è¦ç´ ã‚’ä¸¡æ–¹å‡ºåŠ›ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†
# å‡ºåŠ›ã™ã‚‹å€¤ã¯æ–‡å­—åˆ—ã§ã‚‚æ•´æ•°å€¤ã§ã‚‚æµ®å‹•å°æ•°å€¤ã§ã‚‚ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’å…¥ã‚Œã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

for index, japan_chiho in enumerate(japan):
    print(index, japan_chiho)
```

```python
# 4. ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã§foræ–‡ã‚’ä½¿ã£ã¦ã¿ã¾ã—ã‚‡ã†
# å‡ºåŠ›ã™ã‚‹å€¤ã¯æ–‡å­—åˆ—ã§ã‚‚æ•´æ•°å€¤ã§ã‚‚æµ®å‹•å°æ•°å€¤ã§ã‚‚ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’å…¥ã‚Œã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

print([f"{c}åœ°æ–¹" for c in japan])
```

## ğŸŒ€ ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 2ã€€whileæ–‡
whileæ–‡ã®ä½¿ã„æ–¹ã‚’ç¿’å¾—ã—ã¾ã—ã‚‡ã†
- ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™
- åŒã˜æ–‡ã‚’foræ–‡ã§æ›¸ã

```python
# ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™
count = 0
while count < 4:
    print(count) # 0, 1, 2, 3
    count += 1
```

```python
# åŒã˜æ–‡ã‚’foræ–‡ã§æ›¸ã
for count in range(4):
    print(count) # 0, 1, 2, 3
```

## ğŸŒˆ ç·´ç¿’ãƒ—ãƒ­ã‚°ãƒ©ãƒ 2 (10ç‚¹)
1ï¸âƒ£ ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’whileæ–‡ã§ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚é¡Œæã¯ä½•ã§ã‚‚è‰¯ã„ã§ã™ã€‚    
2ï¸âƒ£ whileæ–‡ã¨åŒã˜å‡¦ç†ã‚’ã€ä»Šåº¦ã¯foræ–‡ã§æ›¸ã„ã¦ã¿ã¾ã—ã‚‡ã†  

```python
# 1. ç‰¹å®šã®å›æ•°ç¹°ã‚Šè¿”ã™ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’whileæ–‡ã§ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†
# å‡ºåŠ›ã™ã‚‹å€¤ã¯ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’å…¥ã‚Œã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

odd = 1
while odd < 20:
    print(odd)
    odd += 2
```

```python
# 2. whileæ–‡ã¨åŒã˜å‡¦ç†ã‚’ã€ä»Šåº¦ã¯foræ–‡ã§æ›¸ã„ã¦ã¿ã¾ã—ã‚‡ã†
# å‡ºåŠ›ã™ã‚‹å€¤ã¯ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’å…¥ã‚Œã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

for odd in range(1,20,2):
    print(odd)
```
