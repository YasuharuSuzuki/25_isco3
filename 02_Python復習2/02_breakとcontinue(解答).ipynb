## ğŸ’” ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 3ã€€breakæ–‡
breakæ–‡ã®ä½¿ã„æ–¹ã‚’ç¿’å¾—ã—ã¾ã—ã‚‡ã†
- ç‰¹å®šã®æ¡ä»¶ã§ãƒ«ãƒ¼ãƒ—ã‚’æŠœã‘ã‚‹

```python
# ç‰¹å®šã®æ¡ä»¶ã§ãƒ«ãƒ¼ãƒ—ã‚’æŠœã‘ã‚‹
list1 = ["a", "b", "c", "d", "e"] # ä»»æ„ã®é…åˆ—
for index, item in enumerate(list1):
    print(item) # a, b, c, d
    if index >= 3:
        break
```

## â˜• ç·´ç¿’ãƒ—ãƒ­ã‚°ãƒ©ãƒ 3 (5ç‚¹)
1ï¸âƒ£ breakã‚’ä½¿ã£ã¦ç‰¹å®šã®æ¡ä»¶ã§ãƒ«ãƒ¼ãƒ—ã‚’æŠœã‘ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚é¡Œæã¯ä½•ã§ã‚‚è‰¯ã„ã§ã™ã€‚   

```python
# 1. ç‰¹å®šã®æ¡ä»¶ã§ãƒ«ãƒ¼ãƒ—ã‚’æŠœã‘ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†
# ä½¿ç”¨ã™ã‚‹å€¤ã¯æ–‡å­—åˆ—ã§ã‚‚æ•´æ•°å€¤ã§ã‚‚æµ®å‹•å°æ•°å€¤ã§ã‚‚ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’ä½¿ç”¨ã—ã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†
japan = ["åŒ—æµ·é“","æ±åŒ—","é–¢æ±","ä¸­éƒ¨","è¿‘ç•¿","ä¸­å›½","å››å›½","ä¹å·"]

for japan_chiho in japan:
    if "åŒ—" not in japan_chiho:
        break
    print(japan_chiho)
```

## ğŸ”œ ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ 4ã€€continueæ–‡
continueæ–‡ã®ä½¿ã„æ–¹ã‚’ç¿’å¾—ã—ã¾ã—ã‚‡ã†
- ç‰¹å®šã®æ¡ä»¶ã§æ¬¡ã®ç¹°ã‚Šè¿”ã—å‡¦ç†ã¸é·ç§»ã™ã‚‹

```python
# ç‰¹å®šã®æ¡ä»¶ã§æ¬¡ã®ç¹°ã‚Šè¿”ã—å‡¦ç†ã¸é·ç§»ã™ã‚‹
list1 = ["a", "b", "c", "d", "e"] # ä»»æ„ã®é…åˆ—
for index, item in enumerate(list1):
    if (index % 2) > 0:
        continue
    print(item) # a, c, e
```

## â­ ç·´ç¿’ãƒ—ãƒ­ã‚°ãƒ©ãƒ 4 (5ç‚¹)
1ï¸âƒ£ continueã‚’ä½¿ã£ã¦ã€ç‰¹å®šã®æ¡ä»¶ã§æ¬¡ã®ç¹°ã‚Šè¿”ã—å‡¦ç†ã¸é·ç§»ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†ã€‚é¡Œæã¯ä½•ã§ã‚‚è‰¯ã„ã§ã™ã€‚   

```python
# 1. ç‰¹å®šã®æ¡ä»¶ã§æ¬¡ã®ç¹°ã‚Šè¿”ã—å‡¦ç†ã¸é·ç§»ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã—ã¦ã¿ã¾ã—ã‚‡ã†
# ä½¿ç”¨ã™ã‚‹å€¤ã¯æ–‡å­—åˆ—ã§ã‚‚æ•´æ•°å€¤ã§ã‚‚æµ®å‹•å°æ•°å€¤ã§ã‚‚ä½•ã§ã‚‚è‰¯ã„ã§ã™ãŒã€ã‚µãƒ³ãƒ—ãƒ«ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã¨ã¯é•ã†ç‹¬è‡ªã®å€¤ã‚’ä½¿ç”¨ã—ã¾ã—ã‚‡ã†
# å¤‰æ•°åã‚‚ç‹¬è‡ªã®ã‚‚ã®ã«å¤‰ãˆã¾ã—ã‚‡ã†

check = "åŒ—æ±ä¸­å›½"

for japan_chiho in japan:
    check_result = [check_char for check_char in check if check_char in japan_chiho]
    if len(check_result) == 0:
        print(f"{japan_chiho}ã«ã¯[" + ",".join(check) + "]ãŒã„ãšã‚Œã‚‚å«ã¾ã‚Œã¦ã„ã¾ã›ã‚“")
        continue

    print(f"{japan_chiho}ã«ã¯[" + "]ã¨[".join(check_result) + "]ãŒå«ã¾ã‚Œã¦ã„ã¾ã™")
```
